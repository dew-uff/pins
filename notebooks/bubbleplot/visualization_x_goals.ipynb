```python
import os, sys, re
from functools import partial
from collections import Counter

import pylab
import pandas as pd
import numpy as np
import seaborn as sns

from pylab import rcParams
from IPython.display import HTML

from bubbleplot import bubble_plot, script_bubble_plot, font0

sys.path.insert(1, os.path.join(sys.path[0], '..', '..'))

import database
from database.groups import YES, NO
from snowballing.operations import load_work, load_citations, reload
from snowballing.approaches import get_approaches, name, wlatex_name, wcitea
reload()

name = lambda x: x.display.replace("  ", "")
line_break = lambda x: '\n'.join(x.split())

all_approaches = get_approaches()
script = [(a, m) for a, m in all_approaches if not m["binary"]]
binary = [(a, m) for a, m in all_approaches if m["binary"]]
len(script)
```

```python
%matplotlib notebook
bubble_plot = partial(bubble_plot, script, binary)
script_bubble_plot = partial(script_bubble_plot, script)
rcParams['figure.figsize'] = 16, 8
sns.set(style='ticks', palette='BuGn') #Set2
```

```python
from database.groups import CACHE, COMPREHENSION, FRAMEWORK, MANAGEMENT, REPRODUCIBILITY, SECURITY
from database.groups import LOG_VIEW, DATA_VIEW, PROPRIETARY, INTEROPERABLE, PROCESS_VIEW, COMBINED_VIEW
from database.groups import INTERNAL, EXTERNAL, CLUSTERING, FILTERING


columns = [CACHE, COMPREHENSION, FRAMEWORK, MANAGEMENT, REPRODUCIBILITY, SECURITY, "", "Overall"]
rows = ["Internal", "External", "Log", "Process", "Data", "Combined", "Clustering", "Filtering", "", "Overall"]
rows = [line_break(x) for x in rows]

places = {
    INTERNAL: "Internal",
    EXTERNAL: "External"
}

types = {
    LOG_VIEW: "Log",
    PROCESS_VIEW: "Process",
    DATA_VIEW: "Data",
    COMBINED_VIEW: "Combined",
}

summarizations = {
    CLUSTERING: "Clustering",
    FILTERING: "Filtering",
}

def process(approach, meta, add):
    def subprocess(elements, categories):
        for element in elements:
            if element in categories:
                for goal in meta["supports"]:
                    add(goal, categories[element], approach)
                    add(goal, "Overall", approach)
                add("Overall", categories[element], approach)
            else:
                print(element, name(approach))
            add("Overall", "Overall", approach)
    subprocess(meta["visplace"], places)
    subprocess(meta["visualization"], types)
    subprocess(meta["summarization"], summarizations)


rcParams['figure.figsize'] = 16, 6
plt, groups = bubble_plot(rows, columns, process, rate=200, label_y=-4, axis_x="← Visualization →")
```

```python
scolumns = [CACHE, COMPREHENSION, FRAMEWORK, MANAGEMENT, REPRODUCIBILITY, "", "Overall"]
rcParams['figure.figsize'] = 15, 3.8
plt, groups = script_bubble_plot(
    rows, scolumns, process, label_y=-5, rate=80, pointsize=15, yticksize=16, xticksize=15,
    axis_x="      Place                                                       Type                                            Summarization",
)
pylab.text(0, -2 , "Visualization", color="black", fontsize=14,
                   horizontalalignment="center",
                   verticalalignment="center")
plt.subplots_adjust(left=0.03, right=1, top=1, bottom=0.2)

plt.show()
plt.savefig("../../../csur/figs/visualization_x_goals.pdf")
```
